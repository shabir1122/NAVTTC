```python
a=[1,2,4,5,7]
```

```python
b=[1,4,16,25,50]
```

```python
plt.plot(a,b)
plt.title("Covid 19 Positivity Ratio")
plt.xlabel("Population")
plt.ylabel("Positivity")
```

```python
c=[1,2,6,8,99]
d=[66,44,33,77,88]
style.use("ggplot")
plt.plot(a,b,"orange",label='line1')
plt.plot(c,d,"b",label='line2')
plt.title("Covid 19 Positivity Ratio")
plt.xlabel('Population')
plt.ylabel('Positivity')
plt.legend()
plt.grid(True,color='k')
```

```python
plt.bar(a,b,color='green')
plt.title("Covid 19 Positivity Ratio")
plt.xlabel('Population')
plt.ylabel('Positivity')
plt.legend()
plt.grid(True,color='k')
```
